```python
def TowerOfHanoi(n, dari, menuju, bantuan):
    if n == 0:
        return
    TowerOfHanoi(n-1, dari, bantuan, menuju)
    print("Pindah disk", n, "dari", dari, "ke", menuju)
    TowerOfHanoi(n-1, bantuan, menuju, dari)

# Driver code
N = 3
# A, C, B are the name of rods
TowerOfHanoi(N, 'A', 'C', 'B')
```

```python
def sortStack(input):
    tmpStack = []

    while len(input) > 0:
        # Pop out the first element from the input stack
        tmp = input[-1]
        input.pop()

        # While the temporary stack is not empty and the top of the stack is smaller than temp
        while len(tmpStack) > 0 and tmpStack[-1] < tmp:
            # Pop from the temporary stack and append it to the input stack
            input.append(tmpStack[-1])
            tmpStack.pop()

        # Append temp to the temporary stack
        tmpStack.append(tmp)

    return tmpStack

# Example usage:
input_stack = [34, 3, 31, 98, 92, 23]
sorted_stack = sortStack(input_stack)
print(sorted_stack)
```

```python
def sortStack(input_stack):
    tmpStack = []

    while len(input_stack) > 0:
        tmp = input_stack[-1]
        input_stack.pop()

        while len(tmpStack) > 0 and tmpStack[-1] < tmp:
            input_stack.append(tmpStack[-1])
            tmpStack.pop()

        tmpStack.append(tmp)

    return tmpStack

def sortArrayUsingStacks(arr, n):
    # Append array elements to the stack
    input_stack = []
    i = 0
    while i < n:
        input_stack.append(arr[i])
        i = i + 1

    # Sort the temporary stack
    tmpStack = sortStack(input_stack)

    i = 0
    # Put stack elements back in arr[]
    while i < n:
        arr[i] = tmpStack[-1]
        tmpStack.pop()
        i = i + 1

# Example usage:
arr = [34, 3, 31, 98, 92, 23]
n = len(arr)

sortArrayUsingStacks(arr, n)
print(arr)
```
